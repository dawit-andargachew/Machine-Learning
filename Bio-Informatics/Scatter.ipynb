```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
data = pd.read_csv('data.csv')

# select only these 3 columns
essential_feature = ['Strain', 'Treatment', 'DT inflection']
data = pd.DataFrame(data, columns=essential_feature)

# drop NA rows
data.dropna(inplace=True)
print(data.shape)

# treatments
# treatments = data['Treatment'].drop_duplicates()
# print(treatments.shape)

# make the order as we want or use the above extracted from data
treatments = ['1% DMSO', '2.5% DMSO', '6.67% H2O', '100 uM AZD8055', '100 uM GDC0941','2.5 ng/mL RAPA', '20 ng/mL RAPA', '1 uM Torin 1', '25 uM Torin 1', '1 mg/mL Caffeine']

# select strains
strain_list = data['Strain'].drop_duplicates()
print(strain_list.shape)
```

```python
# method to make scattering
def scatter_method(DT_inflection,treatment,strain_name):
    # plt.scatter( x=Strain['DT inflection'],y=Strain['Treatment'],marker='.', color='r',s=40)
    plt.scatter(x=DT_inflection, y=treatment, marker='.', color='r', s=40)
    # Fontsize of tick labels
    # plt.xticks(fontsize=5)
    # plt.yticks(fontsize=5)

    # Labels
    plt.ylabel("Treatments")
    plt.xlabel("DT inflection")
    plt.title("Strain: " + strain_name)

    # Display chart
    plt.show()
```

```python
# increase the size to get more
sample = strain_list.head(10)

for s in sample:
    d = data[data['Strain'] == s]

    dt_inflection = d['DT inflection']
    treatment = d['Treatment']

    scatter_method(dt_inflection, treatment, s)
    print('\n')
```
